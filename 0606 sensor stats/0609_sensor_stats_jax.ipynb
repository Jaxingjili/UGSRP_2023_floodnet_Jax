```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {'figure.figsize':(18,10)})
from scipy.ndimage import gaussian_filter1d
```

```python
df = pd.read_excel('sensorstats.xlsx')
df.head()
```

```python
lis = ['Std Deviantion', 'Error', 'Percentage Error']
use = df[['Ground Truth', 'Sensor']].copy()
for name in lis:
    use[name] = gaussian_filter1d(df[name], sigma=2)
use.head()
df = use
```

## Line graph for precentage error

```python
sns.lineplot(data=df, x = 'Ground Truth', y = 'Percentage Error', hue = 'Sensor', lw=3, palette=sns.color_palette("tab10")[:9])
plt.savefig('precent_error')
```

## Graph for errors

```python
sns.lineplot(data=df, x = 'Ground Truth', y = 'Error', hue = 'Sensor', lw=3, palette=sns.color_palette("tab10")[:9])
plt.savefig('error')
```

# Graph for std deviation

```python
sns.lineplot(data=df, x = 'Ground Truth', y = 'Std Deviantion', hue = 'Sensor', lw=3, palette=sns.color_palette("tab10")[:9])
plt.savefig('std deviantion')
```
