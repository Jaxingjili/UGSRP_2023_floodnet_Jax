```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
sns.set(rc = {'figure.figsize':(18,10)})
```

```python
df = pd.read_excel('crtical_angle.xlsx')[2:]
```

```python
df.head()
```

```python
df['Unnamed: 0'] = [i + 1204 for i in df['Unnamed: 0']]
for i in range(1,5):
    df[f'Unnamed: {i}'] += 1
df
```

```python
sns.lineplot(data = df[9:], x = 'Unnamed: 0', y = 'Unnamed: 1', label = 'sensor 3')
sns.lineplot(data = df[9:], x = 'Unnamed: 0', y = 'Unnamed: 2', label = 'sensor 4')
sns.lineplot(data = df[9:], x = 'Unnamed: 0', y = 'Unnamed: 3', label = 'sensor 7')
sns.lineplot(data = df[9:], x = 'Unnamed: 0', y = 'Unnamed: 4', label = 'sensor 10')
plt.savefig('zoom_in_angle')
```

```python
sns.lineplot(data = df, x = 'Unnamed: 0', y = 'Unnamed: 1', label = 'sensor 3')
sns.lineplot(data = df, x = 'Unnamed: 0', y = 'Unnamed: 2', label = 'sensor 4')
sns.lineplot(data = df, x = 'Unnamed: 0', y = 'Unnamed: 3', label = 'sensor 7')
sns.lineplot(data = df, x = 'Unnamed: 0', y = 'Unnamed: 4', label = 'sensor 10')
plt.savefig('whole_angle')
```

```python
df.head()
```

```python
lis = [0,3,4,7,10]
use = df[['Unnamed: 0']].copy()
for i in range(1,5):
    use[f'Unnamed: {i}'] = [math.tan(i) for i in math.pi * (df[f'Unnamed: {i}']/360)] * use['Unnamed: 0']
    sns.lineplot(data = use, x = 'Unnamed: 0', y = f'Unnamed: {i}', label = f'sensor {lis[i]}')
    # sns.lineplot(data = use[9:], x = 'Unnamed: 0', y = f'Unnamed: {i}', label = f'sensor {lis[i]}')
# plt.savefig('zoom_in_tan')
plt.savefig('whole_tan')
```

```python
lis = [0,3,4,7,10]
use = df[['Unnamed: 0']].copy()
for i in range(1,5):
    use[f'Unnamed: {i}'] = [math.tan(i) for i in math.pi * (df[f'Unnamed: {i}']/180)] * use['Unnamed: 0']
    sns.lineplot(data = use, x = 'Unnamed: 0', y = f'Unnamed: {i}', label = f'sensor {lis[i]}')
    # sns.lineplot(data = use[9:], x = 'Unnamed: 0', y = f'Unnamed: {i}', label = f'sensor {lis[i]}')
# plt.savefig('zoom_in_tan')
plt.savefig('whole_angle_tan')
```

```python
lis = [0,3,4,7,10]
use = df[['Unnamed: 0']].copy()
use2 = df[['Unnamed: 0']].copy()
for i in range(1,5):
    use[f'Unnamed: {i}'] = [math.tan(i) for i in math.pi * (df[f'Unnamed: {i}']/360)] * use['Unnamed: 0']
    sns.lineplot(data = use, x = 'Unnamed: 0', y = f'Unnamed: {i}', label = f'sensor {lis[i]}_halved')
    use2[f'Unnamed: {i}'] = [math.tan(i) for i in math.pi * (df[f'Unnamed: {i}']/180)] * use2['Unnamed: 0']
    sns.lineplot(data = use2, x = 'Unnamed: 0', y = f'Unnamed: {i}', label = f'sensor {lis[i]}')
    # sns.lineplot(data = use[9:], x = 'Unnamed: 0', y = f'Unnamed: {i}', label = f'sensor {lis[i]}')
# plt.savefig('zoom_in_tan')
plt.savefig('together_tan')
```
